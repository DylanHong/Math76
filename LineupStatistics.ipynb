```python
#Get the Lakers team_id
from nba_api.stats.static import teams

nba_teams = teams.get_teams()

# Select the dictionary for the Pacers, which contains their team ID
lakers = [team for team in nba_teams if team['abbreviation'] == 'LAL'][0]
lakers_id = lakers['id']
print(f'lakers_id: {lakers_id}')
```

```python
# Query for the last regular season game where the Lakers were playing
from nba_api.stats.endpoints import leaguegamefinder
from nba_api.stats.library.parameters import Season
from nba_api.stats.library.parameters import SeasonType

gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=lakers_id,
                            season_nullable=Season.default,
                            season_type_nullable=SeasonType.regular)  

games_dict = gamefinder.get_normalized_dict()
games = games_dict['LeagueGameFinderResults']
game = games[0]
game_id = game['GAME_ID']
game_matchup = game['MATCHUP']

print("here")

print(f'Searching through {len(games)} game(s) for the game_id of {game_id} where {game_matchup}')
```

```python
#Create Chemistry Score based on individual attributes - needs to be significant
from nba_api.stats.endpoints import commonplayerinfo
trial2 = commonplayerinfo.CommonPlayerInfo('2544').get_data_frames()[1]
trial2.head()

from itertools import islice
LIMIT = 30


# iterrows and unpacking
#for idx, data in islice(df.iterrows(), LIMIT):

#Loop through data frame and grab individual stats for players in the lineup
rand3['CHEM_SCORE'] = np.nan
for index, row in islice(rand3.iterrows(),LIMIT):
    playerids = row['GROUP_ID'].split("-")[1:-1]
    print(playerids)
    teamscore = 0
    for player in playerids:
        playerstats = commonplayerinfo.CommonPlayerInfo(player).get_data_frames()[1]
        try:
            playerscore = playerstats['PTS'] + 2*playerstats['AST'] + 2*playerstats['REB'] + 20*playerstats['PIE']
            print(playerscore)
            teamscore = teamscore + playerscore
        except:
            teamscore = teamscore + 0
            continue
        time.sleep(2)
        
    rand3.at[index,'CHEM_SCORE'] = teamscore
    
rand3.head()
```
